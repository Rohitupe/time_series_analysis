```python
final_df["Date_of_Journey"] = pd.to_datetime(final_df["Date_of_Journey"]).dt.date
```

```python
final_df["Date"] = pd.to_datetime(final_df["Date_of_Journey"]).dt.day
final_df["Month"] = pd.to_datetime(final_df["Date_of_Journey"]).dt.month
final_df["Year"] = pd.to_datetime(final_df["Date_of_Journey"]).dt.year
```

```python
final_df.drop('Date_of_Journey', axis=1, inplace=True)
final_df.head(1)
```

```python
final_df["Arrival_Time"] = final_df["Arrival_Time"].str.split(' ', n=1).str[0]
```

```python
final_df.isnull().sum()
```

```python
final_df["Arrival_Hours"] = final_df["Arrival_Time"].str.split(":").str[0].astype(int)
final_df["Arrival_Mins"] = final_df["Arrival_Time"].str.split(":").str[1].astype(int)
```

```python
final_df.drop("Arrival_Time", axis=1, inplace=True)
```

```python
final_df["Dep_Hours"] = final_df["Dep_Time"].str.split(":").str[0].astype(int)
final_df["Dep_Mins"] = final_df["Dep_Time"].str.split(":").str[1].astype(int)
final_df.drop("Dep_Time", axis=1, inplace=True)
```
